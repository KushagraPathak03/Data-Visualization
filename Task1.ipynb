```python
import pandas as pd
```

```python
data = pd.read_csv('/content/Top10_Countries_GDP.csv')
```

```python
data
```

```python
print(data.dtypes)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Bar Chart: Counts Country per Continent
sns.countplot(x='Continent', color='orange', edgecolor='black', data=data)
plt.title('Count of Countries per Continent')
plt.grid(False)
plt.show()
```

```python
# Barplot: GDP by Country
plt.figure(figsize=(7.5,5))
sns.barplot(y='Country', x='GDP (Trillion USD)', palette='Set1', hue='Country', edgecolor="black", data=data)
plt.grid(True, linestyle='--', axis='x', color='grey')
plt.title('GDP by Country')
plt.show()
```

```python
# Pie Chart: Literacy Rate(%) per Country
plt.figure(figsize=(8, 5))
plt.pie(data['Literacy Rate (%)'], labels=data['Country'], wedgeprops=dict(width=0.8), autopct='%1.1f%%')
plt.title('Literacy Rate(%) per Country')
plt.show()
```

```python
# Lineplot: Education Expenditure vs Literacy Rate
sns.lineplot(x="Education Expenditure (%)", y="Literacy Rate (%)", data=data, color="red")
plt.title('Education Expenditure(%) vs Literacy Rate(%)')
plt.show()
```

```python
#Pointplot: Average Income per Country
sns.pointplot(x='Country', y='Average Income (USD)', data=data, color="purple")
plt.title('Average Income per Country')
plt.xticks(rotation=75)
plt.show()
```

```python
# Boxplot: Population Distribution by Continent
plt.figure(figsize=(7.5, 5))
sns.boxplot(data=data, hue='Continent', y='Population')
plt.title('Population Distribution by Continent')
plt.xlabel('Continent')
plt.legend(title='Continent')
plt.show()
```

```python
# Histogram: Urban Population Distribution
sns.histplot(data['Urban Population (%)'],bins=5, kde=True, color='green', edgecolor='black')
plt.title('Urban Population (%)')
plt.xlabel('Urban Population (%)')
plt.ylabel('Country Count')
plt.axvline(data['Urban Population (%)'].mean(), color='red', linestyle='--', label='Mean Urban Population (%)')
plt.legend()
plt.show()
```

```python
# Scatterplot: Gender Ratio vs Polulation by Country
plt.figure(figsize=(7.5, 5))
sns.scatterplot(data=data, y='Population', x='Gender Ratio (M/F)', color='blue')
plt.title('Gender Ratio vs Polulation')
plt.xlabel('Gender Ratio')
plt.ylabel('Population')
plt.show()
```

```python
# Heatmap: Categorical vs. Numerical Aggregation
pivot_data = data.groupby('Country')[['GDP (Trillion USD)','Literacy Rate (%)', 'Median Age', 'Life Expectancy' ]].mean()
sns.heatmap(pivot_data, annot=True, cmap='YlGnBu')
plt.xticks(rotation=30)
plt.show()
```
